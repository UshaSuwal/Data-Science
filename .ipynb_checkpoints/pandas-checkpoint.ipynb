```python
# 2.List inside list to csv
import csv
data=[['Name','Age','Add'],
      ['Ram',34,'Pkr'],
      ['Hari',66,'Ktm']]
file=open('list_file.csv','w')    # opening file
obj=csv.writer(file)              # object through csv write garna
obj.writerows(data)               # data write gareko file ma
file.close()

df=pd.read_csv('list_file.csv')
df
```

```python
#### DictReader ####

# String data to dictionary

import csv
file=open('stringData.csv')
reader=csv.DictReader(file)
for row in reader:
    print(row)
file.close()
```

```python
# Dictionary in list

import csv
data=[]
file=open('stringData.csv')
reader=csv.DictReader(file)
for row in reader:
    data.append(row)
file.close()
data
```

```python
# Concatination of dictionary and Data Frame

data1={'Name':['Jane','Harry','Rosy','John'],    # dictionary list form
     'Age':[22,34,45,66],
     'Add':['Ktm','Nepal','Japan','China']}
df1=pd.DataFrame(data1)

data2={'Name':['Zane','Parry','Sosy','Kohn'],    # dictionary list form
     'Age':[22,34,45,66],
     'Add':['India','Mars','Moon','Jupiter']}
df2=pd.DataFrame(data2)

df=pd.concat([df1,df2])
df
```
