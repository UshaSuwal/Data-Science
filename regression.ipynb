```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('salary.csv',usecols = ['Years of Experience','Salary'])    # salary and year of experience matra chaiyekole duita matra liyeko
dataset = dataset.dropna()    #empty data removing by removing complete row
dataset
```

```python
# dataset[['Years of Experience']].values
```

```python
# dataset['Years of Experience']
```

```python
X = dataset.iloc[:, 0:-1].values      # taking years of experience values only.   list inside list format ma chaiyekole yesari liyeko otherwise mathy ko tarika le ni lina milxa.
X                                     # 0 to <-1
# X is independent variable.
```

```python
Y = dataset.iloc[:,1:].values       # Taking salary only
Y
# Y is dependent variable
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 1)      

# train_test_split is function
# train size-> min 2/3   60% min
# test size-> max 1/3   40%  max
# Test data size 0.1 means 10% for test dataset and remaining for train data.  Split process will be randomize
# random state is for suffle.
```

```python
print(X_train)
```

```python
# Checking length
total_len = (len(X))
print("All total length = ",total_len)    # total row
print("X_train length and Y_train length = ", len(X_train),len(Y_train))   #total train dataset length
print("X_test length = ", len(X_test))  # total test dataset length
```

```python
#### Training model

# model is imported and not created aafaile.
# model le pattern detect gardai janxa and train hunxa.

from sklearn.linear_model import LinearRegression
model = LinearRegression()    #model
model.fit(X_train,Y_train)
```

```python
### Testing model
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])
print(exp)

# # predict the salary
predicted_salary = model.predict(exp)[0][0]   # list in list  format ko data xa. yo format bata value matra nikalna [0][0] rakheko.
# print(predicted_salary)
print("Your predicted_salary = ",predicted_salary)
```
